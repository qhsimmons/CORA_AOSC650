```python
#Data Collection & Feature Construction

def filter_storms(hurdat, threshold_km):
    #Read in storms, find closest approach, keep only storms under some threshold
    #Output Storm_times dictionary
    
    Storm_times_dict = {}
    for storm_id in hurdat.data.keys():
        storm = hurdat.get_storm(storm_id)

        if storm.year < 1979 or storm.year > 2022:
            continue
            
        df = storm.to_dataframe()

        # Skip empty storms
        if df.empty:
            continue

        # Compute distance to OBX for all timesteps
        dists = haversine(df["lat"].values, df["lon"].values, centroid_lat, centroid_lon)

        min_idx = np.argmin(dists)
        min_dist = dists[min_idx]

        if min_dist < threshold_km:
            closest_time = df["time"].iloc[min_idx]

            start_time = closest_time - timedelta(hours=72)
            end_time = closest_time + timedelta(hours=6)

            Storm_times_dict[storm_id] = (start_time, end_time)
                
        #find closest
        #if closest < 300km, tot_storms += 1, valid_storms_dict[storm] = [closest_approach_time_minus72hours, closest_approach_time_plus6hrs]

    return Storm_times_dict

def get_bathymetry(ds_unstructured, ds_structured, node_coords):

    #Native Mesh
    points = np.column_stack((ds_unstructured["lon"].values, ds_unstructured["lat"].values))
    bathy_vals = ds_unstructured["depth"].values

    #structured grid
    target_points = np.column_stack((ds_structured["lon"].values, ds_structured["lat"].values))

    # Interpolate
    bathy_interp_vals = griddata(
        points,
        bathy_vals,
        target_points,
        method="linear"
    )
    
    bathy_interp = xr.DataArray(
        bathy_interp_vals,
        dims=(["nodes"]),
        coords={"nodes": ds_structured["nodes"]}
    )

    return bathy_interp

def get_spatial_subset(ds):
    x_subset = ((ds.lon >= westLon) & (ds.lon <= eastLon)).compute()
    y_subset = ((ds.lat >= southLat) & (ds.lat <= northLat)).compute()

    boolmask = (x_subset & y_subset).compute()
    nodeIndex = np.where(boolmask)[0]

    ds_subset = ds.isel(nodes=nodeIndex)

    return ds_subset

def get_zeta_stddevs(ds, Storm_times):
    #Get just the data for OBX
    # x_subset = ((ds.lon >= westLon) & (ds.lon <= eastLon)).compute()
    # y_subset = ((ds.lat >= southLat) & (ds.lat <= northLat)).compute()
    
    # boolmask = (x_subset & y_subset).compute()
    # nodeIndex = np.where(boolmask)[0]
    
    zeta_list = []
    
    #Get water elevation for storm(s), right now this will redefine zeta_subset instead of appending
    for storm, times in Storm_times.items():
        start_t, end_t = Storm_times[storm]
        time_subset = ds["zeta"].sel(time=slice(start_t, end_t))
        time_subset = time_subset.expand_dims(storm=[storm])
        zeta_list.append(time_subset)
    
    zeta_subset = xr.concat(zeta_list, dim="storm")
    zeta_subset_6hr = zeta_subset.sel(time=zeta_subset['time'].dt.hour.isin([0, 6, 12, 18])) # Matches 6 hour pattern of hurdat data
    
    
    # ds_500m_subset = ds_500m.isel(nodes=nodeIndex)
    

    # --- coords ---
    lat = np.deg2rad(zeta_subset_6hr.lat.values)
    lon = np.deg2rad(zeta_subset_6hr.lon.values)
    coords = np.vstack([lat, lon]).T
    
    tree = BallTree(coords, metric='haversine')
    r = 0.8 / 6371.0  # 0.8 km in radians
    
    neighbors = tree.query_radius(coords, r=r)
    
    std_list = []
    
    # --- data ---
    for storm_it in range(zeta_subset_6hr.sizes["storm"]):
        zeta = zeta_subset_6hr.isel(storm=storm_it).values  # (nodes, time)
        n_nodes, n_time = zeta.shape #No need to iterate over this each time, but this probably works
        std_feature = np.full((n_nodes, n_time), np.nan)
        
        # --- main loop ---
        for i in range(n_nodes):
            nbrs = neighbors[i]
        
            for t in range(n_time):
        
                # skip if current node is dry (NaN)
                if np.isnan(zeta[i, t]):
                    continue
        
                # keep only valid (wet) neighbors
                valid_mask = ~np.isnan(zeta[nbrs, t])
                valid_nbrs = nbrs[valid_mask]

                #Only calculate for nodes with more than 5 wet neighbors (Doesn't this do less than 5?)
                if len(valid_nbrs) < 5:
                    continue
        
                std_feature[i, t] = np.std(zeta[valid_nbrs, t])
    
        std_da = xr.DataArray(
            std_feature,
            dims=["node", "time"],
            coords={
                "node": np.arange(n_nodes),
                "time": zeta_subset_6hr["time"]
            },
            name="std_dev"
        )
    
        std_list.append(std_da)
    
    std_all = xr.concat(std_list, dim="storm")
    std_all = std_all.assign_coords(storm=zeta_subset_6hr["storm"])
    std_all = std_all.transpose("storm", "time", "node")

    return std_all, std_all["node"], zeta_subset_6hr
    

def get_storm_features(node_coords, Storm_times):
    storm_list = []

    for storm, (start, end) in Storm_times.items():
        storm_ = hurdat.get_storm((storm, start.year))
        ds_storm_ = storm_.to_xarray()
        ds_storm_ = ds_storm_.rename({'lat': 'hur_lat', 'lon': 'hur_lon'})
        ds_storm_6hr_ = ds_storm_.sel(time=ds_storm_['time'].dt.hour.isin([0, 6, 12, 18])).sel(time=slice(start, end))
    
        ds_storm_6hr_ = ds_storm_6hr_.expand_dims(storm=[storm])
    
        storm_list.append(ds_storm_6hr_)
    
    ds_storm_all = xr.concat(storm_list, dim="storm")
    ds_storm_all = ds_storm_all.expand_dims(node=node_coords)
    ds_storm_all = ds_storm_all.transpose("storm", "time", "node")

    return ds_storm_all

def combine_features(std_all, ds_storm_all, ds_depths_subset, zeta_subset_6hr, node_coords):
    combined_ds = xr.merge([std_all, ds_storm_all])
    combined_ds["zeta_std_dev"] = std_all
    combined_ds["bathy"] = ("node", ds_depths_subset.values)
    combined_ds["lat"] = ("node", zeta_subset_6hr["lat"].values)
    combined_ds["lon"] = ("node", zeta_subset_6hr["lon"].values)
    combined_ds = combined_ds.drop_vars(["extra_obs", "special"])

    return combined_ds

def azimuth(lat1, lon1, lat2, lon2):

    lat1 = np.radians(lat1)
    lon1 = np.radians(lon1)
    lat2 = np.radians(lat2)
    lon2 = np.radians(lon2)

    az = np.arctan2(np.sin(lon2-lon1)*np.cos(lat2), np.cos(lat1)*np.sin(lat2) - np.sin(lat1)*np.cos(lat2)*np.cos(lon2-lon1))
    
    return az

def haversine(lat1, lon1, lat2, lon2):

    lat1 = np.radians(lat1)
    lon1 = np.radians(lon1)
    lat2 = np.radians(lat2)
    lon2 = np.radians(lon2)

    #R is defined in constants, R = 6371 km
    a = np.sin((lat2-lat1)/2)**2 + np.cos(lat1)*np.cos(lat2)*np.sin((lon2-lon1)/2)**2
    c = 2*np.arctan2(np.sqrt(a), np.sqrt(1-a))
    d = R*c
    
    return d
```

```python
#Feature Addition

def add_features(combined_ds):

    dist = haversine(combined_ds["hur_lat"], combined_ds["hur_lon"], combined_ds["lat"], combined_ds["lon"])
    az = azimuth(combined_ds["hur_lat"], combined_ds["hur_lon"], combined_ds["lat"], combined_ds["lon"])
    
    combined_ds["dist_to_hur"] = dist
    combined_ds["azimuth"] = az
    combined_ds["dx"] = np.sin(combined_ds["azimuth"])
    combined_ds["dy"] = np.cos(combined_ds["azimuth"])
    combined_ds["dx_dist_weighted"] = combined_ds["dist_to_hur"] * combined_ds["dx"]
    combined_ds["dy_dist_weighted"] = combined_ds["dist_to_hur"] * combined_ds["dy"]
    
    return combined_ds

#May want to add an extra 6 hours at the beginning of each storm so I can drop the NaN and have a true 72 hr reading
def hur_speed(ds):
    lat1 = ds["hur_lat"]
    lon1 = ds["hur_lon"]

    lat0 = lat1.shift(time=1)
    lon0 = lon1.shift(time=1)

    dist = haversine(lat0, lon0, lat1, lon1)
    
    x_dist = haversine(lat0, lon0, lat0, lon1)
    y_dist = haversine(lat0, lon0, lat1, lon0)
    
    speed = dist / 6.0
    x_speed = x_dist / 6.0
    y_speed = y_dist / 6.0

    # speed = speed.expand_dims(nodes=ds["nodes"]).transpose("storm", "time", "nodes")

    ds["hur_speed"] = speed
    ds["hur_x_speed"] = x_speed
    ds["hur_y_speed"] = y_speed

    #Double check this
    cos_theta = (x_dist * x_speed + y_dist * y_speed) / (dist * speed + 1e-6)
    sin_theta = (x_dist * y_speed - y_dist * x_speed) / (dist * speed + 1e-6)

    ds["alignment"] = cos_theta
    ds["cross_track"] = sin_theta

    return ds

def intensification(ds):
    vmax1 = ds["vmax"]
    mslp1 = ds["mslp"]

    vmax0 = vmax1.shift(time=1)
    mslp0 = mslp1.shift(time=1)

    d_vmax_dt = (vmax1 - vmax0) / 6.0
    d_mslp_dt = (mslp1 - mslp0) / 6.0

    ds["d_vmax_dt"] = d_vmax_dt
    ds["d_mslp_dt"] = d_mslp_dt

    return ds
```
