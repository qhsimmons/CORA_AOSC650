```python
#Imports from PlotMesh
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import dask
import intake
import xarray as xr
import holoviews as hv
import datashader as ds
import datashader.transfer_functions as tf
import datashader.utils as du
import geoviews as gv
from holoviews import opts
import holoviews.operation.datashader as dshade
from holoviews.operation.datashader import datashade, shade, rasterize
import cmocean
hv.extension('bokeh')

#Imports from BoundingBox
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from cartopy.mpl.gridliner import Gridliner, LongitudeFormatter, LatitudeFormatter
import contextily as cx

#Imports for Tropycal
import tropycal.tracks as tracks
```

```python
#Constants

#Box Edges
westLon, eastLon, northLat, southLat = -77.262338, -75.367403, 36.553085, 34.552033

#Storm Times
Irene_start_t, Irene_end_t = "2011-08-24 23:00:00", "2011-08-27 23:00:00"
```

```python
catalog = intake.open_catalog("s3://noaa-nos-cora-pds/CORA_V1.1_intake.yml",storage_options={'anon':True})
# list(catalog)
```

```python
#Pull full data sets

ds = catalog["CORA-V1.1-fort.63-timeseries"].to_dask()
ds_grid = catalog["CORA-V1.1-Grid-timeseries"].to_dask()

#Fix lat lon labels in ds_grid (natively flipped)
# ds_grid = ds_grid.rename({"lon": "lat", "lat": "lon"})
```

```python
ds_grid
```

```python
#Get just the data for OBX
x_subset = ((ds_grid.lon >= westLon) & (ds_grid.lon <= eastLon)).compute()
y_subset = ((ds_grid.lat >= southLat) & (ds_grid.lat <= northLat)).compute()

boolmask = (x_subset & y_subset).compute()
nodeIndex = np.where(boolmask)[0]

#Get water elevation for storm(s)
start_t, end_t = Irene_start_t, Irene_end_t

zeta_subset = ds["zeta"].sel(time=slice(start_t, end_t)).isel(node=nodeIndex)
```

```python

#Get mesh for OBX
ds_subset = ds.isel(node=nodeIndex)
v = np.vstack((ds_subset.x, ds_subset.y, ds_subset.depth)).T
verts = pd.DataFrame(v, columns=['x', 'y', 'z'])

# --- Build OBX triangles correctly ---
# Full connectivity (global)
elem_raw = ds["element"].to_numpy().astype(int)

# Convert element indexing to 0-based global node POSITIONS
nnode = ds.sizes["node"]
emin, emax = elem_raw.min(), elem_raw.max()

if emin == 1 and emax <= nnode:
    elem0 = elem_raw - 1          # 1-based -> 0-based
elif emin == 0 and emax < nnode:
    elem0 = elem_raw              # already 0-based
else:
    raise ValueError(f"Unexpected element indexing: min={emin}, max={emax}, nnode={nnode}")

# Keep only triangles fully inside the OBX nodeIndex set
sel = np.asarray(nodeIndex, dtype=int)
mask = np.isin(elem0, sel).all(axis=1)
elem_sub = elem0[mask]

if elem_sub.size == 0:
    raise ValueError("No triangles fully inside OBX mask. Check bbox / ds_grid lon-lat orientation.")

# Remap global node positions -> local 0..(n_obx-1)
global_to_local = {g: i for i, g in enumerate(sel)}
elem_local = np.vectorize(global_to_local.get, otypes=[int])(elem_sub)

tris = pd.DataFrame(elem_local, columns=["v0", "v1", "v2"])

# Sanity check (prevents your current out-of-bounds crash)
nverts = len(verts)
tri_max = tris.to_numpy().max()
assert tri_max < nverts, f"Connectivity still out of bounds: max={tri_max}, nverts={nverts}"

points = gv.operation.project_points(gv.Points(verts, vdims=['z']))
tris = pd.DataFrame(ds_subset['element'].values.astype('int')-1, columns=['v0','v1','v2'])
```

```python
#Print mesh
opts.defaults(
    opts.Image(width=1000, height=700),
    opts.RGB(width=1000, height=700))
wireframe = datashade(hv.TriMesh((tris,verts)).edgepaths)
trimesh = hv.TriMesh((tris,hv.Points(verts, vdims='z')))
dmesh = dshade.rasterize(trimesh).opts(cmap=cmocean.cm.deep, colorbar=True)
dmesh * wireframe
```

```python
import datashader as dsh
from holoviews.operation.datashader import rasterize, datashade

nodes = hv.Points(verts, kdims=["x","y"], vdims=["z"])
trimesh = hv.TriMesh((tris, nodes))

wireframe = datashade(trimesh.edgepaths)
dmesh = rasterize(trimesh, aggregator=dsh.mean("z")).opts(
    cmap=cmocean.cm.deep, colorbar=True, width=1000, height=700
)

dmesh * wireframe
```
