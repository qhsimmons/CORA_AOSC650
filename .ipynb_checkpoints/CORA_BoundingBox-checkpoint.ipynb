# This notebook allow users to select NOAA Coastal Ocean ReAnalysis (CORA) data by bounding box. The data is stored on NOAA's Open Data Dissemination (NODD) service. The data will be loaded as an xarray dataset. For a use case example, water level data is extracted at a bounding box near Myrtle Beach during Hurricane Hugo in 1989. The data is scatter plotted over a basemap tile and then a time series is plotted at one of the nodes.

## To access data on the NODD with this notebook, you must be sure you are using a version of zarr < 3.0.0. If your current installation of zarr is version 3, please pip install zarr==2.18.7.

```python

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import dask
import intake
import xarray as xr
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from cartopy.mpl.gridliner import Gridliner, LongitudeFormatter, LatitudeFormatter
import contextily as cx
```

**Access the data on the NODD and initialize the available CORA datasets.** 

*This accesses a .yml file located on the NODD that shows which CORA output files are available to import.*

```python

catalog = intake.open_catalog("s3://noaa-nos-cora-pds/CORA_V1.1_intake.yml",storage_options={'anon':True})
list(catalog)
```

**CORA-V1.1-fort.63:** Hourly water levels-- **'zeta'** is the water elevation variable <br>
**CORA-V1.1-swan_DIR.63:** Hourly mean wave direction-- **'swan_DIR'** is the mean wave direction variable <br>
**CORA-V1.1-swan_TPS.63:** Hourly peak wave periods-- **'swan_TPS'** is the peak wave period variable <br>
**CORA-V1.1-swan_HS.63:** Hourly significant wave heights-- **'swan_HS'** is the significant wave height variable <br>
**CORA-V1.1-Grid:** Hourly water levels interpolated from model nodes to 500-meter resolution coastal grid-- **'zeta'** is the water elevation variable<br>
>*All datasets denoted as **'-timeseries'** are optimized for pulling long time series (greater than a few days) <br>
For up to a few days of data, use the regular dataset (not labeled **'-timeseries'** in the catalog description)*

***Now, create an xarray dataset for the CORA data that you would like to use.***

```python
ds = catalog["CORA-V1.1-fort.63-timeseries"].to_dask()
```

```python
ds
```

**Select data at model nodes within a bounding box.**

```python
minX, maxX, minY, maxY = -78.9, -78.85, 33.65, 33.7

x_subset =  ((ds.x >= minX) & (ds.x <= maxX)).compute()
y_subset = ((ds.y >= minY) & (ds.y <= maxY)).compute() 

boolmask  = (x_subset & y_subset).compute()
nodeIndex = np.where(boolmask)[0] 
print(nodeIndex)

start_t = "1989-09-22 00:00:00"
end_t = "1989-09-22 23:00:00"
zeta_subset = ds['zeta'].sel(time=slice(start_t, end_t)).isel(node=nodeIndex)
```

**Scatter plot the data over a basemap.**

```python
# Extract the data and coordinate arrays for plotting
lons = zeta_subset['x'].values
lats = zeta_subset['y'].values

projection = ccrs.PlateCarree()

# Create a figure and a GeoAxes object
fig = plt.figure(figsize=(10, 8))

# Set the projection for the map. `PlateCarree` is a common choice for plotting lat/lon data.
ax = fig.add_subplot(1, 1, 1, projection=projection)

# Add basemap features
buf=0
ax.set_extent([minX-buf, maxX+buf, minY-buf, maxY+buf], crs=projection) # Set the extent to focus on Europe
# ax.add_feature(cfeature.LAND)
# ax.add_feature(cfeature.COASTLINE)
# ax.add_feature(cfeature.BORDERS, linestyle=':')

# Add a basemap tile using contextily
cx.add_basemap(ax, crs=projection, source=cx.providers.Esri.WorldGrayCanvas)

# Scatter the points onto the map
# Use the `transform` argument to specify the CRS of the scatter points.
time=5
vmin=zeta_subset.min(skipna=True).values
vmax=zeta_subset.max(skipna=True).values
data=zeta_subset.isel(time=time).values

scatter = ax.scatter(
    lons,
    lats,
    c=data,
    cmap='RdBu_r',
    s=50,    # Size of the points
    transform=projection,
    vmin=vmin,
    vmax=vmax
)

# Call ax.gridlines() and store the returned Gridliner object
gl = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True,
                  linewidth=1, color='gray', alpha=0.5, linestyle='--')

# Configure which labels to display
gl.top_labels = False
gl.right_labels = False

# 3. Set the formatters for decimal degrees
gl.xformatter = LongitudeFormatter(dms=False)
gl.yformatter = LatitudeFormatter(dms=False)

# Optional: Adjust label styles
gl.xlabel_style = {'size': 10, 'color': 'black'}
gl.ylabel_style = {'size': 10, 'color': 'black'}

# Add labels
ax.set_title("Nodes in Bounding Box")

# Add a colorbar for the scattered data
cbar = plt.colorbar(scatter, ax=ax, label="Water Level (m)")

# Add node number labels.
# This loop is effective for a small number of points.
for i in range(len(nodeIndex)):
    # Annotate with the node number. `xy` is the point to label.
    # `xytext` is the position of the label text.
    ax.annotate(
        f"{nodeIndex[i]}",
        xy=(lons[i], lats[i]),
        xytext=(3, 3), # Offset the text slightly
        textcoords='offset points',
        transform=ccrs.PlateCarree(),
        fontsize=8,
        ha='left'
    )
    
# Display the plot
plt.show()
```
